# 함수 세상의 모든 것을 입력과 출력으로 바라보기

머신러닝 : 입력과 출력이 있을 때 숨어있는 입력과 출력의 관계를 찾는 과정

함수의 정의
- 입력과 출력의 관계를 나타내는 것 
f: x -> y x는 정의역 y는 공역

```python
import numpy as np
import matplotlib.pyplot as plt

# y= x^2
x = np.linspace(-10, 10, 100)
# y = x ** 2
y = pow(x, 2)
y2 = pow(x+1, 2)

plt.plot(x, y, label='y = x^2')
plt.plot(x, y2, label='y = (x+1)^2')

# x축과 y축 추가 및 화살표 표시
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# 화살표 추가 (x축)
plt.annotate('', xy=(10, 0), xytext=(0, 0),
             arrowprops=dict(arrowstyle="->", color='black'))

# 화살표 추가 (y축)
plt.annotate('', xy=(0, 100), xytext=(0, 0),
             arrowprops=dict(arrowstyle="->", color='black'))

# x축 간격을 1로 맞추기
plt.xticks(np.arange(-10, 11, 1))

# y축 범위를 100으로 제한
plt.ylim(-10, 100)

plt.grid(True)
plt.title('Plot of y = x^2 and y = (x+1)^2')
plt.legend()
plt.show()
```

합의 기호

$$
\sum_{k=1}^{n} a_k = a_1 + a_2 + \cdots + a_{n-2} + a_{n-1} + a_n
$$

1에서 부터 n 까지 a_k에 대입해서 전개해서 더하는 것 

$$
\sum_{k=1}^{n} (a_k + b_k) = \sum_{k=1}^{n} a_k + \sum_{k=1}^{n} b_k
$$
$$
\sum_{k=1}^{n} (a_k - b_k) = \sum_{k=1}^{n} a_k - \sum_{k=1}^{n} b_k
$$
$$
\sum_{k=1}^{n} c \cdot a_k = c \sum_{k=1}^{n} a_k
$$
$$
\sum_{k=1}^{n} c = c \cdot n
$$

$$
\sum_{i=1}^{n} \sum_{j=1}^{m} a_i b_j
$$

- 역함수 


로지스틱 시그모이드 함수


```python
import numpy as np
import matplotlib.pyplot as plt

# 로지스틱 시그모이드 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# x 값 범위 설정
x = np.linspace(-10, 10, 100)

# 시그모이드 함수 적용
y = sigmoid(x)

# 결과 그래프로 나타내기
plt.plot(x, y)
plt.title('Logistic Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```
