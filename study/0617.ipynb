문제 IQR 방법으로 이상치 검출 하는 예제

1. IQR 방법을 사용하여 이상값을 탐지합니다.
2. 데이터와 이상값을 시각화하여 이상값이 잘 탐지되었는지 확인합니다.

```python
import pandas as pd
import numpy as np

# 랜덤 데이터셋 생성
np.random.seed(0)
data = pd.DataFrame({
    'value': np.random.normal(50, 10, 100)  # 평균 50, 표준편차 10인 정규분포 데이터 100개 생성
})

# 일부 데이터에 이상값 추가
data.loc[::10, 'value'] = np.random.normal(100, 1, 10)  # 평균 100, 표준편차 1인 데이터 10개 추가

```

```python

# IQR 계산
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# 이상값 경계 계산
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

print("**이상값 하한 및 상한값 **")
print("이상값 최소치 :", lower_bound)
print("이상값 최대치 :", upper_bound)

# 이상값 식별
outliers = [value for value in data if value < lower_bound or value > upper_bound]

print("이상값:", outliers)
```

```python
import pandas as pd
import numpy as np

# 랜덤 데이터셋 생성
np.random.seed(0)
data = pd.DataFrame({
    'value': np.random.normal(50, 10, 100)  # 평균 50, 표준편차 10인 정규분포 데이터 100개 생성
})

# 일부 데이터에 이상값 추가
data.loc[::10, 'value'] = np.random.normal(100, 1, 10)  # 평균 100, 표준편차 1인 데이터 10개 추가

# 숫자 데이터만 선별
numeric_data = data[data['value'].apply(lambda x: isinstance(x, (int, float)))]

# IQR 계산
q1 = np.percentile(numeric_data['value'], 25)
q3 = np.percentile(numeric_data['value'], 75)
iqr = q3 - q1

# 이상값 경계 계산
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

print("**이상값 하한 및 상한값 **")
print("이상값 최소치 :", lower_bound)
print("이상값 최대치 :", upper_bound)

# 이상값 식별
outliers = numeric_data[(numeric_data['value'] < lower_bound) | (numeric_data['value'] > upper_bound)]['value'].tolist()

print("이상값:", outliers)
```
