1.5 막대 그래프 그려보기

-폰트 설정

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

plt.bar(hat['hatchery'],hat['chick'])
plt.show()
```

그래프 색상추가

```python
plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'],hat['chick'],
        color = ('red','orange','yellow','green','blue','navy','purple'))
plt.title('부화장별 병아리 수')
plt.xlabel('가로')
plt.ylabel('세로')
```

seaborn 패키지로 팔레트

```python
import seaborn as sns
col7 = sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'],hat['chick'],
        color = col7,edgecolor='black')
plt.title('부화장별 병아리 수')
plt.xlabel('가로')
plt.ylabel('세로')
```

1-7 그래프 위에 택스트 추가하기

```python
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]+0.5,y[i],ha='center')
```

```python
col7 = sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'],hat['chick'],
        color = col7,edgecolor='black')
addtext(hat['hatchery'],hat['chick'])
plt.title('부화장별 병아리 수')
plt.xlabel('가로')
plt.ylabel('세로')

plt.show()
```

1-8 그래프에 선그리기

빨간색 수평선 추가

```python
col7 = sns.color_palette('Pastel2',7)

plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'],hat['chick'],
        color = col7,edgecolor='black')
addtext(hat['hatchery'],hat['chick'])




plt.hlines(30,-1,7,color = 'red', linestyles='dashed')



plt.title('부화장별 병아리 수')
plt.xlabel('가로')
plt.ylabel('세로')
plt.rcParams.update({'font.size':12})
plt.show()
```

1-9 파이차트 그려보기

파이차트 -> 항목별 비율 확인

```python
# 파이차트를 그리기 위해 비율 계산

pct = hat['chick']/hat['chick'].sum()
pct
```

```python
col7 = sns.color_palette('Pastel2',7)
plt.figure(figsize=(10,10))
plt.pie(pct,labels=hat['hatchery'],autopct='%lf%%',colors = col7,startangle=90,counterclock=False)
plt.show()
```
