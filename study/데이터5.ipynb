그리고 병아리들의 몸무게를 측정했는데 병아리들의 몸무게는 얼마일까?

데이터 불러와서 구조 유형 확인

```python
import pandas as pd
b = pd.read_csv('ch4-2.csv') # 변수 b에 데이터셋 입력
b.head()
# chick_nm : 이름 weight : 몸무게
```

```python
b.info() # b 데이터셋 정보 확인
```

통계량으로 분포 확인하기

```python
b.describe() # b 데이터 셋 기초 통계량 확인
# 추가적으로 1사 분위수(25%), 3사 분위수(75%)
```

표준편차

만약에 부화장 A,B 몸무게 평균은 38g 동일   
표준편차가 크면 병아리를 사육하는데 문제가 발생   
큰 병아리들 때문에 몸무게 왜소한 병아리들이 먹이를 빼앗길수 있음

히스토그램으로 분포 확인하기

평균 최소값 표준편차 최댓값 등은 숫자로만 데이터 확인  -> 그래프로 확인

```python
# 그래프 한글 깨짐 문제 해결을 위해서 맑은 고딕 폰트 지정
from matplotlib import font_manager, rc
font_path = 'c:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family = font_name)
```

```python
import matplotlib.pyplot as plt

# 히스토그램 그리기
plt.figure(figsize=(10,7))
plt.hist(b.weight, bins=7)
plt.title('B 부화장 병아리 무게 분포 현황', fontsize = 17)
plt.xlabel('병아리무게(g)')
plt.ylabel('마릿수')
plt.show()
```

상자그림 분포 확인하기

```python
# 상자그림 그리기
plt.figure(figsize=(8,10))
plt.boxplot(b.weight)
plt.title('B 부화장 병아리 무게 상자 그림', fontsize = 17)
plt.ylabel('병아리무게(g)')
plt.show()
```

다중그래프 분포 확인

히스토그램 + 상자그림을 동시에 그림

```python
import matplotlib.pyplot as plt

# 히스토그램과 상자 한번에 그리기
plt.figure(figsize=(10,12))
# 달라지는 부분
plt.subplot(2,1,1)
plt.hist(b.weight, bins=7)
plt.title('B 부화장 병아리 무게 분포 현황', fontsize = 17)
plt.subplot(2,1,2)
plt.boxplot(b.weight, vert=False)
plt.show()
```

중심 극한 정리 - 데이터가 많을 경우(일반적으로 30개 이상) 정규분포와 가까워진다는 것을 정리한 것이 중심극한정리이다.
대략적인 분포 알수 있다
평균 u, 표준편차를 a

최소값 u - 3*a
1사 분위수 u - 0.645*a
3사 분위수 u + 0.645*a
최대값 u + 3*a


```python
import pandas as pd
import matplotlib.pyplot as plt

# 부화장 B 병아리 무게 분포 시각화
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# 히스토그램
ax1.hist(b.weight, color='skyblue', edgecolor='black')
ax1.set_xlabel('병아리 무게 (g)')
ax1.set_ylabel('빈도')
ax1.set_title('B 부화장 병아리 무게 분포 현황')

# 박스플롯
ax2.boxplot(b.weight, vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
ax2.set_xlabel('병아리 무게 (g)')
ax2.set_title('B 부화장 병아리 무게 분포')

plt.tight_layout()
plt.show()

```
